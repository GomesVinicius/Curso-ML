## Apagando todos outliers

```python
Q1 = np.percentile(df['concavity_se'], 25)
Q3 = np.percentile(df['concavity_se'], 75)
Q3
```

```python
C = 1.5
IIQ = Q3 - Q1
LI = Q1 - C*IIQ
LS = Q3 + C*IIQ

print(IIQ)
print(LI)
print(LS)
```

```python
columns = df.drop('diagnosis', axis=1).columns.values
columns
```

```python
C = 1.5
df_sem_outlier = df.copy()
print(df_sem_outlier.shape)

for x in columns:
  Q1 = np.percentile(df[x], 25)
  Q3 = np.percentile(df[x], 75)

  IIQ = Q3 - Q1
  LI = Q1 - C * IIQ
  LS = Q3 + C * IIQ

  query = '`{}` <= {} | `{}` >= {}'.format(x, LI, x, LS)
  ids = df.query(query)

  df_sem_outlier.drop(index=ids.index, axis=0, inplace=True, errors='ignore')

print(df_sem_outlier.shape) #(547, 31)
```
